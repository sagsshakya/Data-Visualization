## A normal distribution from a binomial distribution

- Let's take a coin and flip it. The probability of getting a head or a tail is 50%. If you
take the same coin and flip it six times, the probability of getting a head three times
can be computed using the following formula:

    - P(x) = nCx (p)^x (q)^(n-x)
    
        - where, x = number of successes desired and x = 0,1,2,...,n.
        - n = total number of trials.
        - p = probabilty of success in a trial.
        - q = probabilty of failure in a trial.
        - p + q = 1

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
n = 6
p = 0.5   #50%
x = np.arange(n+1)
rv = binom(n,p)
y = rv.pmf(x)

#plotting the binomial distribution.
plt.vlines(x,y, ymax=0, color = 'b', lw = 2)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

#### Increasing the number of trials.

```python
n = 500
p = 0.5   #50%
x = np.arange(n+1)
rv = binom(n,p)
y = rv.pmf(x)

#plotting the binomial distribution.
plt.vlines(x,y, ymax=0, color = 'brown', lw = 2)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.show()
```

#### Changing p - value.

```python
n = 500
p1 = 0.6   #60%
p2 = 0.7
p3 = 0.8

x = np.arange(n+1)

#for p1
rv1 = binom(n,p1)
y1 = rv1.pmf(x)

#for p2
rv2 = binom(n,p2)
y2 = rv2.pmf(x)

#for p3
rv3 = binom(n,p3)
y3 = rv3.pmf(x)

#plotting the binomial distribution.
fig = plt.figure(figsize=(18,9))

#for p = 0.6
plt.subplot(1,3,1)
plt.vlines(x,y1,ymax=0, colors='b', lw = 2, label = 'p = 0.6')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('p = 0.6')

#for p = 0.7
plt.subplot(1,3,2)
plt.vlines(x,y2,ymax=0, colors='r', lw = 2)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('p = 0.7')

#for p = 0.8
plt.subplot(1,3,3)
plt.vlines(x,y3,ymax=0, colors='g', lw = 2)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('p = 0.8')

plt.show()
```

## Z - Score

- A z-score, in simple terms, is a score that expresses the value of a distribution in
standard deviation with respect to the mean. 
    -Calculating the z - score:
            - z = (x - mean)/ s.d

- A classroom has 100 students in it and they have just got their mathematics
examination score. We simulate the score of these 60 students with a normal
distribution.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

score = np.random.normal(50, 1.5, 100).round(2)
z = stats.zscore(score)

plt.figure(figsize = (18,10))

plt.subplot(1,2,1)
plt.hist(x=score,bins=20, color='red', normed = True)
plt.xlabel('Class Scores')
plt.ylabel('Frequency')
plt.title('X - value histogram')

plt.subplot(1,2,2)
plt.hist(z, bins=20, color='b', normed = True)
plt.xlabel('Z - Scores')
plt.ylabel('Frequency')
plt.title('Z - value histogram')


plt.show()
```

### Notice that the x - graph is more or less symmetrical about the mean i.e 50 whereas the z - score normalized distribution is more or less symmetrical about the standard mean i.e 0.
